```python
"""394. Decode String"""
class Solution:
    def decodeString(self, s: str) -> str:
        stack=list()

        for i in range(len(s)):
            c=s[i]
            if c != "]":
                stack.append(c)
            else:                
                str_buffr=""
                while stack[-1]!= "[":
                    str_buffr = stack.pop() + str_buffr
                stack.pop()

                num_buffr=""
                while stack and stack[-1].isnumeric():
                    num_buffr = stack.pop()+num_buffr
                stack.append(str_buffr*int(num_buffr))
        return("".join(stack))
```

```python
"""2514. Count Anagrams
"""

from collections import defaultdict
from functools import reduce
from math import factorial

def mult(a , b):
  return a * b

class Solution:
    def countAnagrams(self, s: str) -> int:
        l=s.split()

        each_word_permutations_list=list()

        for word in l:
          m = defaultdict(int)
          for c in word:
              m[c]+=1
          permu_count=factorial(len(word))//reduce(mult,[factorial(v) for v in m.values()])
          each_word_permutations_list.append(permu_count)

        val=reduce(mult, [t for t in each_word_permutations_list])
        val=int(val%(10**9 + 7))
        return(val)
```
