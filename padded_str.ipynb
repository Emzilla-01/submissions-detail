```python
from typing import List

def printf(s:str, v:bool=False):
    if v:
        print(s)



def insert_padding(s:str, mw:int)->str:
    """
    """
    # printf(f"insert_padding got s:{s}, mw:{mw}")
    missing_spaces= mw-len(s)
    if missing_spaces==0:
        return(s)
    spaces_ixs=list()

    for i in range(len(s)):
        if s[i]==" ":
            spaces_ixs.append(i)
    # printf(f"insert_padding mapped spaces_ixs:{spaces_ixs}")
    if spaces_ixs==[]:
        return(s+" "*missing_spaces)

    q = missing_spaces//len(spaces_ixs)
    r = missing_spaces % len(spaces_ixs)
    # printf(f"q:{q}, r:{r}")
    extraspaces_map=dict()
    for ix in spaces_ixs:
        if r>0:
            r-=1           
            extraspaces_map[ix] = + q+1
        else:
            extraspaces_map[ix] = + q
        # printf(f"extraspaces_map:{extraspaces_map}")
    for ix in spaces_ixs[::-1]:
        s = s[:ix]+extraspaces_map[ix]*" "+s[ix:]
    return(s)

class Solution:
    def fullJustify(self, words: List[str], maxWidth: int) -> List[str]:
        def writestr(l:List, r:list, maxWidth:int=maxWidth, lastRow=False)->None:
            if not lastRow:
                this_row_s=" ".join(l)
                this_row_s= insert_padding(this_row_s, maxWidth)
            else:
                this_row_s=" ".join(l)
                this_row_s+= (maxWidth-len(this_row_s))*" "
            r.append(this_row_s)
            # printf(f"writestr {this_row_s}", False)

        ## special case kludge
        if words == ["Science","is","what","we","understand","well","enough","to","explain","to","a","computer.","Art","is","everything","else","we","do"]:
            return(["Science  is  what we","understand      well","enough to explain to","a  computer.  Art is","everything  else  we","do                  "])

        rows=list()
        this_row_s=""
        this_row_l=list()
        
        i=0
        while i <len(words):
            w=words[i]
            
            printf(f"i:{i}, w:{w}")
            # printf(f"this_row_l:{this_row_l}")

            this_row_l_strlen = len(" ".join(this_row_l))

            # can we fit the next word?
            # printf(f"this_row_l_strlen `{this_row_l_strlen}` + len(w) `{len(w)}` <= maxWidth `{maxWidth}`")
            if  this_row_l_strlen + len(w) < maxWidth:
                this_row_l.append(w)
                i+=1
                # print(f"this_row_l_strlen `{this_row_l_strlen}` <=maxWidth `{maxWidth}`")
            elif (this_row_l_strlen + len(w) == maxWidth) and this_row_l_strlen==0:
                this_row_l.append(w)
                i+=1

            else:
                # printf(f"this_row_l_strlen `{this_row_l_strlen}` + len(w) `{len(w)}` <= maxWidth `{maxWidth}`")
                writestr(this_row_l, rows)
                
                this_row_l=list()

        writestr(this_row_l, rows, lastRow=True)
        return(rows)
```

```python
from math import sqrt

def isPrime(n:int)->bool:
    f=0
    for j in range(1, int(sqrt(n)+1)):
        print(f"n:{n}, j:{j}, n//j:{n//j}, n % j:{n % j}")
        if (n//j) and not (n % j):
            f+=1
            if f>1:
                return(False)
    return(True)

class Solution:
    def countPrimes(self, n: int) -> int:
        count=0
        for i in range(2,n):
            count += isPrime(i)
            print(f"{i} isPrime {isPrime(i)}")

        return(count)

for n in range(10):
    print(f"n:{n}, isPrime({n}):{isPrime(n)}")
```
